# Segmentacion elemento CIELO





```python
# Tomamos una muestra
img_muestra_cielo = imgRGB[0:250, 0:953,:]
plt.figure()
plt.imshow(img_muestra_cielo)
plt.show()
```

```python
# Calculamos el promedio y la dstd de la muestra en los 3 canales
color_mean, color_std = cv.meanStdDev(img_muestra_cielo)
print(color_mean.flatten().tolist())
print(color_std.flatten().tolist())
```

```python
# Pintemos de blanco los puntos que estÃ©n cerca de la media en un radio menor que 3 sigmas
# para Â¨capturarÂ¨ el 99.7 % de pixeles.
n_sigma =2.7
mask = cv.inRange(imgRGB, color_mean-color_std * n_sigma,  color_mean+color_std * n_sigma)
img_segmentada = cv.bitwise_and(imgRGB, imgRGB, mask=mask)

plt.figure()
plt.imshow(img_segmentada)
plt.show()

plt.figure()
plt.imshow(mask,cmap='gray')
plt.show
```

# Segmentacion elemento Agua

```python
# Tomamos una muestra
img_muestra_agua = imgRGB[300:400,0:953,:]
plt.figure()
plt.imshow(img_muestra_agua)
plt.show()
```

```python
# Calculamos el promedio y la dstd de la muestra en los 3 canales
color_mean_agua, color_std_agua = cv.meanStdDev(img_muestra_agua)
print(color_mean.flatten().tolist())
print(color_std.flatten().tolist())
```

```python
# Pintemos de blanco los puntos que estÃ©n cerca de la media en un radio menor que 3 sigmas
# para Â¨capturarÂ¨ el 99.7 % de pixeles.
n_sigma = 0.7
mask_agua = cv.inRange(imgRGB, color_mean_agua-color_std_agua * n_sigma,  color_mean_agua+color_std_agua * n_sigma)
img_segmentada_agua = cv.bitwise_and(imgRGB, imgRGB, mask=mask_agua)

plt.figure()
plt.imshow(img_segmentada_agua)
plt.show()

plt.figure()
plt.imshow(mask_agua,cmap='gray')
plt.show
```

```python
# Pintemos de blanco los puntos que estÃ©n cerca de la media en un radio menor que 3 sigmas
# para Â¨capturarÂ¨ el 99.7 % de pixeles.
n_sigma = 1
mask_Rocas = cv.inRange(imgRGB, color_mean_rocas-color_std_rocas * n_sigma,  color_mean_rocas+color_std_rocas * n_sigma)
img_segmentada_Rocas = cv.bitwise_and(imgRGB, imgRGB, mask=mask_Rocas)

plt.figure()
plt.imshow(img_segmentada_Rocas)
plt.show()

plt.figure()
plt.imshow(mask_Rocas,cmap='gray')
plt.show
```
